Data Visualization

```python
##Lets see rank of each Continent
d = data.groupby(by = "Continent").sum()
continents = d.index
ranks = d['Rank']
plt.figure(figsize=(10,6))
plt.bar(continents , ranks,color="skyblue")
plt.xlabel('Continent')
plt.ylabel('Rank')
plt.show()
```

```python
columns = ['2022 Population','2020 Population','2015 Population','2010 Population','2000 Population','1990 Population','1980 Population','1970 Population']
for year in columns:
    plt.figure(figsize=(10,6))
    plt.plot(continents , d[year])
    plt.title(f"Population for {year}")
    plt.xlabel("Continents")
    plt.ylabel(f'{year}')
    
```
